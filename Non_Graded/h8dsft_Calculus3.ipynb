```python
import numpy as np
import sympy as sy
```

```python
x,y = sy.symbols('x y')
f = 3*x + 4*y - 5
f
```

```python
# turunan fungsi f terhdap x
sy.diff(f, x)
```

```python
# turunan fungsi f terhadap y
sy.diff(f, y)
```

## Nilai

```python
def f(X, Y):
    return 3*X + 4*Y - 5

X = np.linspace(1, 20, 10)
Y = np.linspace(1, 20, 10)


Z = np.array([[f(i, j)for i in X] for j in Y])

dx , dy = np.gradient(Z)
```
