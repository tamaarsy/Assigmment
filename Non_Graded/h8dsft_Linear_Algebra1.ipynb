## 2. Diberikan 3 bauh vector 3 dimensi:
## A = [17,22,19] B = [10,20,11] C = [5,12,9]

```python
A = np.array([17,22,19]) 
B = np.array([10,20,11]) 
C = np.array([5,12,9])
```

```python
# petambahan
A + B
```

```python
#pengurangan
B-C
```

```python
# A dot C
A.dot(C)
```

```python
# A x B
A * B
```

```python
np.linalg.norm(A)
```

```python
result = A.dot(B)
```

```python
#mencari sudut
theta = np.arccos(result/(np.linalg.norm(A) * np.linalg.norm(B)))
theta
```
