## Hitung verktor gradien (turunan pasial) masing masing pixel untuk masing masing sumbu x dan y : 

```python

gambar = plt.imread('gambar.jpg')
```

```python
col = [] # mengambil pixel baris
for i in (range(1, 439)):
    col.append(gambar[i-1, i])
```

```python
len(col)
```

```python
#mengambil pixel colom
bar = []
for a in range(1,781):
    bar.append(gambar[:, i-1:i])
```

```python
len(bar)
```

```python
# memhitung vektor gradien (turunan pasial) masing - masing pixel unuk masing masing sumbu x dan y
list_x = []
list_y = []
def grad(pix):
    if len(pix) == 438: # sumbu y(baris)
        for a in pix:
            for i in a:
                i = np.gradient(i)
                list_y.append(i)
                
    elif len(pix) == 780:# sumbu x(column)
        for a in pix:
            for i in a:
                for o in i:
                    o = np.gradient(o)
                    list_x.append(o)
grad(bar)
grad(col)
```
